# 00 · Ejecutor de Notebooks — TP Final Aprendizaje Automático

Monta Drive **una sola vez** y corre los notebooks `01` → `06` en orden, en el mismo kernel.

**Antes de correr:** verificá que `NOTEBOOKS_DIR` apunte a la carpeta correcta en tu Drive.

```python
# ⚙️  CONFIGURACIÓN — editá solo esta celda si hace falta
from google.colab import drive
drive.mount('/content/drive')

import os

# Carpeta en Drive donde están los notebooks 01_*.ipynb … 06_*.ipynb
NOTEBOOKS_DIR = '/content/drive/MyDrive/TP_Final Aprendizaje Automatico'

print('Directorio:', NOTEBOOKS_DIR)
print('Existe    :', os.path.isdir(NOTEBOOKS_DIR))
```

```python
import glob

# Busca notebooks con prefijo numérico: 01_*.ipynb … 99_*.ipynb
notebooks = sorted(glob.glob(os.path.join(NOTEBOOKS_DIR, '[0-9][0-9]_*.ipynb')))

if not notebooks:
    raise FileNotFoundError(
        f'No se encontraron notebooks [0-9][0-9]_*.ipynb en:\n  {NOTEBOOKS_DIR}\n'
        'Revisá que NOTEBOOKS_DIR sea correcto.'
    )

print(f'Notebooks encontrados ({len(notebooks)}):')
for nb in notebooks:
    print(f'  {os.path.basename(nb)}')
```

```python
import nbformat
from IPython import get_ipython

ip = get_ipython()
results = []

for i, nb_path in enumerate(notebooks, 1):
    name = os.path.basename(nb_path)
    sep  = '=' * 65
    print(f'\n{sep}')
    print(f'[{i}/{len(notebooks)}]  {name}')
    print(f'{sep}\n')

    try:
        with open(nb_path, encoding='utf-8') as f:
            nb = nbformat.read(f, as_version=4)

        for j, cell in enumerate(nb.cells):
            if cell.cell_type != 'code' or not cell.source.strip():
                continue
            res = ip.run_cell(cell.source)
            if res.error_in_exec is not None:
                raise res.error_in_exec

        print(f'\n✓  {name} completado.')
        results.append((name, True, None))

    except Exception as e:
        import traceback
        msg = traceback.format_exc()
        print(f'\n✗  Error en {name}:\n{msg}')
        results.append((name, False, str(e)))
        # Continúa con el siguiente notebook aunque haya error

# ─── Resumen final ───────────────────────────────────────────────
print('\n' + '=' * 65)
print('RESUMEN FINAL')
print('=' * 65)
for name, ok, err in results:
    mark = '✓' if ok else '✗'
    print(f'  {mark}  {name}')
    if err:
        print(f'       → {err[:120]}')

total_ok = sum(1 for _, ok, _ in results if ok)
print(f'\n  {total_ok}/{len(results)} notebooks completados exitosamente.')
```
