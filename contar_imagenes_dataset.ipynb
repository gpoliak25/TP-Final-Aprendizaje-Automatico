Paso 1: montar Google Drive

Paso 2: definir la ruta donde están las carpetas

```python
ruta_base = "/content/drive/MyDrive/TP_Final Aprendizaje Automatico"
```

Paso 3: definir las carpetas a contar

```python
carpeta_ok = ruta_base + "/Imagenes OK"
carpeta_patologicas = ruta_base + "/Imagenes Patologicas"
extensiones_imagen = (
    ".jpg",
    ".jpeg",
    ".png",
    ".bmp",
    ".tif",
    ".tiff",
    ".webp",
    ".dcm",
    ".dicom"
)
```

Paso 4: contar las imágenes

```python
import os
for ruta_actual, subcarpetas, archivos in os.walk(carpeta_patologicas):
    print("CARPETA:", ruta_actual)
    print("SUBCARPETAS:", len(subcarpetas))
    print("ARCHIVOS:", len(archivos))
    print("--------------------")
```

```python
def contar_imagenes(carpeta):
    contador = 0

    for ruta_actual, subcarpetas, archivos in os.walk(carpeta):
        for archivo in archivos:
            if archivo.lower().endswith(extensiones_imagen):
                contador += 1

    return contador
```

```python
cantidad_ok = contar_imagenes(carpeta_ok)
cantidad_patologicas = contar_imagenes(carpeta_patologicas)
```

Paso 5: mostrar el resultado

```python
print("Cantidad de imágenes OK:", cantidad_ok)
print("Cantidad de imágenes Patológicas:", cantidad_patologicas)
print("Total de imágenes:", cantidad_ok + cantidad_patologicas)
```
